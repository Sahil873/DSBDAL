```python
import requests # to make request to webpage
from bs4 import BeautifulSoup # For parsing HTML and XML content
```

```python
URL = "https://www.flipkart.com/realme-gt-master-cosmos-black-128-gb/p/itmcc74e8302084d?pid=MOBG5JPXUXV9PYTT&lid=LSTMOBG5JPXUXV9PYTT2LBUUO&marketplace=FLIPKART&q=realme+gt+master+edition&store=tyy%2F4io&srno=s_1_1&otracker=search&otracker1=search&fm=search-autosuggest&iid=e48c3103-1c9d-42e5-88fa-abbffdf28bcd.MOBG5JPXUXV9PYTT.SEARCH&ppt=sp&ppn=sp&ssid=hpjv9foar40000001715004606584&qH=03b5320058efca24"
response = requests.get(URL) # returns request object
print(response) # prints status code
```

```python
response.content # 
```

```python
soup = BeautifulSoup(response.content)
```

```python
image_link = soup.find('img', class_="_0DkuPH").get('src')
from IPython.display import display, Image
image = requests.get(image_link)
display(Image(image.content))
```

```python
product_name = soup.find('span', class_="VU-ZEz").text.strip()
print(product_name)
```

```python
avg_ratings = soup.find('div', class_="ipqd2A").text.strip()
print(avg_ratings)
```

```python
entire_reviews = soup.findAll('div', class_="col EPCmJX")
for single_review in entire_reviews:
    rating = single_review.find('div', class_="XQDdHH Ga3i8K").text.strip()
    review = single_review.find('div', class_='').text.strip("READ MORE")
    customer = single_review.find('p', class_="_2NsDsF AwS1CA").text.strip()
    print(customer + "\nRated - " + rating + "\nReview - " + review + "\n\n")                                            
```
